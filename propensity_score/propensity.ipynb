```python
#層別マッチング(0.05刻み)
interval = np.arange(0,1.05,0.05)
match_list = []
table=df[["trt","death"]]
table["ps"]=ps
for i in range(0,len(interval)-1):
    temp0 = table[(table['trt']==0) & (interval[i] < table['ps']) & (table['ps'] < interval[i+1])]
    temp1 = table[(table['trt']==1) & (interval[i] < table['ps']) & (table['ps'] < interval[i+1])]
    if (len(temp0) > 0) & (len(temp1) > 0):
        match_list.append(temp1['death'].mean()-temp0['death'].mean())

np.mean(match_list)
```
